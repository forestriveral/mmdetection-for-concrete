```python
import pandas as pd
import numpy as np
import copy
import json
```

```python
data_path = "./original_data.csv"
```

```python
data = pd.read_csv(data_path)
print(len(data.columns))
```

```python
def count_types(heads):
    ts = []
    for head in heads:
        types = []
        for d in data[head]:
            if d not in types:
                types.append(d)
        ts.append(types)
    return ts
```

```python
heads = ["Vehicle type", "weather condition", "status"]

ts = count_types(heads)
ts
```

```python
def instruction(heads, ts):
    ins = {}
    for i, h in enumerate(heads):
        for j, t in enumerate(ts[i]):
            if h not in ins.keys():
                ins[h] = {}
            ins[h][t] = j + 1
    
    with open("./instruction.json", "w+") as f:
        json.dump(ins, f, indent=4)
    
    return ins
```

```python
ins = instruction(heads, ts)
ins
```

```python
def data_vectorization(data, heads, types, save=None):
    data.loc[data["Vehicle type"]=="microbus","Vehicle type"]
    data = copy.deepcopy(data)
    for i, h in enumerate(heads):
        for j in types[i]:
            data.loc[data[h] == j, h] = types[i].index(j) + 1
            
    x = range(5, len(data.columns))
    data.drop(data.columns[x],axis=1,inplace=True)
    shuffled_data = data.sample(frac=1).reset_index(drop=True)
    
    if save:
        shuffled_data.to_csv(save, index=0)
    
    return shuffled_data
```

```python
data = data_vectorization(data, heads, ts, save="./input_data.csv")

data
```
